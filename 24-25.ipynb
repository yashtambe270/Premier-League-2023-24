```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = r"D:\PREMIER LEAGUE 2023-24.xlsx"
xls = pd.ExcelFile(file_path)

# Load the "GENERAL" sheet
df_premier_league = pd.read_excel(xls, sheet_name="GENERAL")

# Set style
sns.set(style="whitegrid")
```

```python
# 1 **Top Scoring Teams**
df_goals = df_premier_league.sort_values("GOALS", ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(y=df_goals["TEAMS"], x=df_goals["GOALS"], palette="coolwarm")
plt.xlabel("Total Goals Scored")
plt.ylabel("Teams")
plt.title("Top Scoring Teams in Premier League 2023-24")
plt.show()
```

```python
# 2 **Best Defensive Teams (Tackles & Clearances)**
df_defense = df_premier_league.sort_values("TACKLES", ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(y=df_defense["TEAMS"], x=df_defense["TACKLES"], palette="Blues")
plt.xlabel("Total Tackles")
plt.ylabel("Teams")
plt.title("Teams with Most Tackles in Premier League 2023-24")
plt.show()
```

```python
# 3 **Most Disciplined Teams (Yellow & Red Cards)**
df_cards = df_premier_league.sort_values("YELLOW CARDS", ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(y=df_cards["TEAMS"], x=df_cards["YELLOW CARDS"], palette="Oranges")
plt.xlabel("Total Yellow Cards")
plt.ylabel("Teams")
plt.title("Teams with Most Yellow Cards in Premier League 2023-24")
plt.show()
```

```python
# 4 **Passing Trends (Through Balls & Long Balls)**
df_passing = df_premier_league.sort_values("THROUGH BALLS", ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(y=df_passing["TEAMS"], x=df_passing["THROUGH BALLS"], palette="Greens")
plt.xlabel("Total Through Balls")
plt.ylabel("Teams")
plt.title("Teams with Most Through Balls in Premier League 2023-24")
plt.show()
```
